## Introduction

We'll perform some basic data analysis using Python in this notebook.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Generation

Let's create some sample data to work with.

```python
# Create a sample DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}

df = pd.DataFrame(data)
df
```

## Data Analysis

Now let's perform some basic analysis on our data.

```python
# Calculate the mean of each column
df.mean()
```

```python
# Perform some calculations
total_sum = df.values.sum()
print(f"Sum of all values: {total_sum}")

total_product = np.prod(df.values)
print(f"Product of all values: {total_product}")
```

## Visualization

Let's create a simple visualization of our data.

```python
# Create a heatmap
plt.figure(figsize=(6, 4))
plt.imshow(df.values, cmap='viridis')
plt.colorbar()
plt.title('Sample Heatmap')
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.index)), df.index)
plt.tight_layout()
plt.show()
```
